## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML beats a fixed rule here

A fixed rule can only use simple thresholds, such as selecting pages with high impressions or old content age. However, page performance depends on multiple interacting factors, including impressions, clicks, CTR, average position, content age, freshness, and engagement metrics.

Machine learning can learn these relationships automatically and identify patterns that are difficult to capture with a single hand-written rule. The goal is to support human decision-making by prioritizing pages for review rather than replacing human judgment.

```python
features = [
    "impressions_90d",
    "clicks_90d",
    "ctr",
    "avg_position",
    "content_age_days",
    "sessions_90d"
]

print("Example ML features:")
for feature in features:
    print("-", feature)
```
